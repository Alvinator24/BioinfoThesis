```python
import joblib
import numpy as np
import pandas as pd

# Load datasets
X_test_TCGA  = pd.read_csv("../Datasets/TCGA Dataset Splits/X_test.csv")
y_test_TCGA  = pd.read_csv("../Datasets/TCGA Dataset Splits/Y_test.csv").values.ravel()

X_test_TCGA_purified = pd.read_csv("../Datasets/TCGA InfiniumPurify Dataset Splits/X_test.csv")
y_test_TCGA_purified = pd.read_csv("../Datasets/TCGA InfiniumPurify Dataset Splits/Y_test.csv").values.ravel()

X_test_Norway = pd.read_csv("../Datasets/Norway Dataset Splits/X_test.csv")
y_test_Norway = pd.read_csv("../Datasets/Norway Dataset Splits/Y_test.csv").values.ravel()

X_test_Norway_purified = pd.read_csv("../Datasets/Norway InfiniumPurify Dataset Splits/X_test.csv")
y_test_Norway_purified = pd.read_csv("../Datasets/Norway InfiniumPurify Dataset Splits/Y_test.csv").values.ravel()

# Load RF models
rf_TCGA = joblib.load("../RandomForest/TCGA No Infinium/PKL/RF_TCGA_SMOTE_RFE.pkl")
rf_TCGA_IP = joblib.load("../RandomForest/TCGA with InfiniumPurify/PKL/RF_TCGA_IP_SMOTE_RFE.pkl")
rf_Norway = joblib.load("../RandomForest/Norway No Infinium/PKL/RF_Norway_NoImb_No_RFE.pkl")
rf_Norway_IP = joblib.load("../RandomForest/Norway with InfiniumPurify/PKL/RF_Norway_IP_IRUS_RFE.pkl")

# Load SVM models
svm_TCGA = joblib.load("../SVM/svm.pkl")
svm_TCGA_IP = joblib.load("../SVM/Purified/svm_rfe.pkl")
svm_Norway = joblib.load("../SVM/svm_smote_Norway.pkl")
svm_Norway_IP = joblib.load("../SVM/Purified/svm_Norway.pkl")

# Save predictions for plotting later
np.save("RF_TCGA_preds.npy", rf_TCGA.predict_proba(X_test_TCGA))
np.save("RF_TCGA_IP_preds.npy", rf_TCGA_IP.predict_proba(X_test_TCGA_purified))
np.save("RF_Norway_preds.npy", rf_Norway.predict_proba(X_test_Norway))
np.save("RF_Norway_IP_preds.npy", rf_Norway_IP.predict_proba(X_test_Norway_purified))

np.save("SVM_TCGA_preds.npy", svm_TCGA.predict_proba(X_test_TCGA))
np.save("SVM_TCGA_IP_preds.npy", svm_TCGA_IP.predict_proba(X_test_TCGA_purified))
np.save("SVM_Norway_preds.npy", svm_Norway.predict_proba(X_test_Norway))
np.save("SVM_Norway_IP_preds.npy", svm_Norway_IP.predict_proba(X_test_Norway_purified))
```

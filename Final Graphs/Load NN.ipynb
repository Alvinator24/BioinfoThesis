```python
import os, gc, warnings
import numpy as np
import pandas as pd
import matplotlib; matplotlib.use('Agg')
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Dropout, Input, BatchNormalization, ReLU, LeakyReLU
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import regularizers
from tensorflow.keras.optimizers import Adam
from scikeras.wrappers import KerasClassifier
from sklearn.metrics import confusion_matrix, classification_report, f1_score
from sklearn.preprocessing import LabelEncoder, label_binarize
from sklearn.metrics import precision_recall_curve, average_precision_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense as _OrigDense

def create_nn(meta, neurons_layer1=256, neurons_layer2=128,
              dropout_rate1=0.3, dropout_rate2=0.3,
              learning_rate=0.002, activation='relu'):
    n_features_in = meta["n_features_in_"]
    n_classes_     = meta["n_classes_"]
    model = Sequential([
        Input(shape=(n_features_in,)),
        Dense(neurons_layer1, activation=activation),
        BatchNormalization(),
        Dropout(dropout_rate1),
        Dense(neurons_layer2, activation=activation),
        BatchNormalization(),
        Dropout(dropout_rate2),
        Dense(n_classes_, activation='softmax'),
    ])
    model.compile(optimizer=Adam(learning_rate=learning_rate),
                  loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Load datasets
X_test_TCGA  = pd.read_csv("Datasets/TCGA Dataset Splits/X_test.csv")
y_test_TCGA  = pd.read_csv("Datasets/TCGA Dataset Splits/Y_test.csv").values.ravel()

X_test_TCGA_purified = pd.read_csv("Datasets/TCGA InfiniumPurify Dataset Splits/X_test.csv")
y_test_TCGA_purified = pd.read_csv("Datasets/TCGA InfiniumPurify Dataset Splits/Y_test.csv").values.ravel()

X_test_Norway = pd.read_csv("Datasets/Norway Dataset Splits/X_test.csv")
y_test_Norway = pd.read_csv("Datasets/Norway Dataset Splits/Y_test.csv").values.ravel()

X_test_Norway_purified = pd.read_csv("Datasets/Norway InfiniumPurify Dataset Splits/X_test.csv")
y_test_Norway_purified = pd.read_csv("Datasets/Norway InfiniumPurify Dataset Splits/Y_test.csv").values.ravel()

# Load NN models
nn_TCGA = joblib.load("NeuralNetworks/NN_RS_bestmodel.pkl")
nn_TCGA_IP = joblib.load("NeuralNetworks/Purified/NN_RS_Purified_bestmodel.pkl")
nn_Norway = joblib.load("NeuralNetworks/NN_Norway_SMOTE_RS_bestmodel.pkl")
nn_Norway_IP = joblib.load("NeuralNetworks/Purified/NN_Norway_SMOTE_RS_Purified_bestmodel.pkl")

# Save predictions
np.save("NN_TCGA_preds.npy", nn_TCGA.predict_proba(X_test_TCGA))
np.save("NN_TCGA_IP_preds.npy", nn_TCGA_IP.predict_proba(X_test_TCGA_purified))
np.save("NN_Norway_preds.npy", nn_Norway.predict_proba(X_test_Norway))
np.save("NN_Norway_IP_preds.npy", nn_Norway_IP.predict_proba(X_test_Norway_purified))
```
